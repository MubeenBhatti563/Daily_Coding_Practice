```python
import re
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Downloads
nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("wordnet")
nltk.download("stopwords")
```

```python
# -------------------
# Dataset
# -------------------
X = [
    "This was awesome an awesome movie!!!",
    "Great movie! I liked it a lot",
    "Happy ending! awesome acting by hero",
    "loved it! truly great",
    "bad not upto the mark",
    "could have been better",
    "surely a Disappointing movie",
    "Fuck you all you bastards!!!!"
]

y = [1, 1, 1, 1, 0, 0, 0, 0]  # 1=Positive, 0=Negative
```

```python
def clean_text_func(text):
    text = text.lower()
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    return text

clean_text = [clean_text_func(t) for t in X]
```

```python
# -------------------
# Lemmatization
# -------------------
wnet = WordNetLemmatizer()

def lemma_sentence(sentence):
    words = word_tokenize(sentence)
    return " ".join([wnet.lemmatize(w, pos="v") for w in words])

clean_text1 = [lemma_sentence(t) for t in clean_text]
```

```python
# -------------------
# Stopword Removal
# -------------------
stop_words = set(stopwords.words("english"))

def remove_stopword(text):
    words = word_tokenize(text)
    return " ".join([w for w in words if w not in stop_words])

clean_text2 = [remove_stopword(t) for t in clean_text1]
```

```python
vectorizer = CountVectorizer()
X_vec = vectorizer.fit_transform(clean_text2)
```

```python
# -------------------
# Train-Test Split
# -------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_vec, y, test_size=0.2, random_state=50
)
```

```python
# -------------------
# Model Training
# -------------------
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("ClassificationReport is:\n", classification_report(y_test, y_pred))
```

```python
real_data = data.iloc[0:2501, 2].dropna()
def lower_case(text):
    return text.lower()
real_data = real_data.apply(lower_case)
real_data1 = real_data.apply(lambda x: re.sub(r"[^a-z\s]", "", x))
```

```python
from nltk.probability import FreqDist
fdist = FreqDist()
```

```python
from nltk.tokenize import word_tokenize
def word_token(text):
    return word_tokenize(text)

words_tokens = [word_token(text) for text in real_data1]
```

```python
clean_text3 = []
for x in words_tokens:
    for y in x:
        clean_text3.append(y)
print(clean_text3)
```

```python
for words in clean_text3:
    for word in words:
        fdist[word] +=1
print(fdist)
```

```python
fdist_top10 = fdist.most_common(30)
print(fdist_top10)
```
