```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
```

```python
df = {
    'text': ["I am so excited for today!", "I feel really bad about the situation.", "I am just here, nothing special.", "This is the best day ever!", "I am feeling really down today.", "I don't care about this anymore."],
    'sentiment': ["Happy", "Sad", "Neutral", "Happy", "Sad", "Neutral"]
}

df = pd.DataFrame(df)
```

```python
df['text'] = df['text'].str.lower().str.replace(r'[^a-zA-Z\s]', '', regex=True)
```

```python
df['tokens'] = df['text'].str.split()
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['text'])
bow_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
```

```python
X = vectorizer.fit_transform(df['text'])
Y = df['sentiment']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, Y,
    test_size=0.2,
    random_state=42
)
```

```python
classifier_nb = MultinomialNB()
classifier_nb.fit(X_train, y_train)

y_pred_nb = classifier_nb.predict(X_test)

print("Naive Bayes Accuracy:", accuracy_score(y_test, y_pred_nb))
print("Naive Bayes Classification Report:\n", classification_report(y_test, y_pred_nb))
```

```python
classifier_lr = LogisticRegression(max_iter=200)  # max_iter needed for convergence on small datasets
classifier_lr.fit(X_train, y_train)

y_pred_lr = classifier_lr.predict(X_test)

print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
print("Logistic Regression Classification Report:\n", classification_report(y_test, y_pred_lr))
```
