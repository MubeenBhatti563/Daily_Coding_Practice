```python
# Importing important libraries
import pandas as pd
import nltk
```

```python
text_data = pd.read_csv("train.csv")
text_data.shape
```

```python
description = text_data["Description"]
description.head()
```

### Converting Description texts to lower case

```python
import re

def to_lower(text):
    text = text.lower()
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    return text

description = description.apply(to_lower)
description.head(5)
```

### Tokenization

```python
nltk.download('punkt_tab')
description.dropna()
description.shape
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
sent_token = []
for word in description:
    sent = sent_tokenize(word)
    sent_token.append(sent)
```

```python
sent_token1 = []
for i in sent_token:
    for j in i:
        sent_token1.append(j)
```

```python
# word Tokenization
word_token = [word_tokenize(word) for word in sent_token1]
```

```python
print(len(word_token))
```

```python
# Removing stop words
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
clean_text = []
for words in word_token:
    w = []
    for word in words:
        if not word in stopwords.words('english'):
            w.append(word)
    clean_text.append(w)
```

```python
from nltk.stem.porter import PorterStemmer
port = PorterStemmer()
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
wnet = WordNetLemmatizer()
```

```python
lem = []
for words in clean_text:
    w = []
    for word in words:
        clean = wnet.lemmatize(word, pos='v')
        w.append(clean)
    lem.append(w)
```
